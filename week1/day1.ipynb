```python
# 1. List operations
stocks = ["AAPL", "GOOGL", "TSLA", "MSFT", "AMZN"]
print(stocks[0])          # first item
print(stocks[-1])         # last item
print(stocks[1:3])        # slicing
stocks.append("NVDA")     # add item
print(len(stocks))        # length
stocks                    # new list
```

```python
# 2. List comprehension (very useful in data science)
prices = [150, 2800, 220, 310, 3400, 450]
above_300 = [p for p in prices if p > 300]
doubled = [p * 2 for p in prices]
print(above_300)
print(doubled)
```

```python
# 3. Dictionaries
portfolio = {"AAPL": 150, "GOOGL": 2800, "TSLA": 220}
print(portfolio["AAPL"])       # access
portfolio["NVDA"] = 450        # add key
for ticker, price in portfolio.items():
    print(f"{ticker}: ${price}")
```

```python
# 4. Functions
def calculate_return(buy_price, sell_price):
    return_pct = ((sell_price - buy_price) / buy_price) * 100
    return round(return_pct, 2)

print(calculate_return(150, 180))   # should print 20.0
print(calculate_return(2800, 2500)) # should print -10.71
```
